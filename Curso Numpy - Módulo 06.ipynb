# Números Aleatórios

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.__version__
```

## Gerador de dados aleatórios

### `np.random.default_rgn`

```python
# define um gerador
rgn = np.random.default_rng(seed=1)
rgn.random()
```

## Dados aleatórios simples

### `rgn.random`

```python
# Gera um número aleatório entre 0 e 1
# diferente a cada execução
rgn.random()
```

```python
# Gera um número aleatório entre 0 e 1
# igual a cada execução
rgn = np.random.default_rng(seed=1)
rgn.random()
```

```python
# Gera um array de 10 números aleatórios
# entre 0 e 1
rgn.random(size = 10)
```

```python
# Gera uma matrix 5x2 de números aleatórios entre 0 e 1
# entre 0 e 1
rgn.random(size = (5, 2))
```

### `rgn.integers`

```python
# retorna 5 números inteiros aleatórios no intervalo entre 0 e 3
rgn.integers(4, size=5)
```

```python
# retorna 5 números inteiros aleatórios no intervalo entre 0 e 49
rgn.integers(low=10, high=50, size=5)
```

```python
# retorna 3 números inteiros aleatórios
# o primeiro entre 10 e 14
# o segundo entre 10 e 53
# o terceiro entre 10 e 99
rgn.integers(low=10, high=[15, 54, 100])
```

### `rgn.choice`

```python
# definindo um array
array = np.arange(0, 10)
array
```

```python
# sorteia aleatoriamente um numero no array
rgn.choice(array)
```

```python
# sorteia aleatoriamente 5 numeros no array
rgn.choice(array, size=5)
```

```python
# sorteia aleatoriamente 10 numeros no array
# o mesmo valor poder ser selecionado mais de uma vez
rgn.choice(array, size=10)
```

```python
# sorteia aleatoriamente 10 numeros no array
# o mesmo valor NÃO poder ser selecionado mais de uma vez
rgn.choice(array, size=10, replace=False)
```

```python
# definindo probabilidades para o sorteio dos elementos
array = np.array([1, 2, 3])

# a soma das probabilidades deve ser igual a 1
p = np.array([0.5, 0.4, 0.1])
```

```python
# chamando a função de sorteio
rgn.choice(array, size=10, p=p)
```

```python
# fazendo um exemplo com matrizes
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# seleciona duas linhas aleatorias do array
rgn.choice(a=array_2d, size=2, axis=0)
```

```python
# seleciona duas colunas aleatorias do array
rgn.choice(a=array_2d, size=2, axis=1)
```

## Permutações

### `rgn.shuffle`

```python
# criando um array simples
array = np.arange(0, 10)
array
```

```python
# embaralha o array de forma aleatoria
rgn.shuffle(array)
array
```

```python
# definindo um array 2d
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# escolhendo o eixo a ser embaralhado (linhas)
rgn.shuffle(array_2d, axis=0)
array_2d
```

```python
# definindo um array 2d
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# escolhendo o eixo a ser embaralhado (colunas)
rgn.shuffle(array_2d, axis=1)
array_2d
```

### `rgn.permutation`

É similar ao shuffle mas gera uma cópia dos dados.

```python
# gera um array embaralhado com números de 0 a 9
# tem o comportamento de chamar:
# primeiro: array = np.arage(10)
# depois: np.shaffle(array)
rgn.permutation(10)
```

```python
# definindo um array 2d
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# escolhendo o eixo a ser embaralhado (linha)
rgn.permutation(array_2d, axis=0)
array_2d
```

```python
# definindo um array 2d
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# escolhendo o eixo a ser embaralhado (colunas)
rgn.permutation(array_2d, axis=1)
array_2d
```

### `rgn.permuted`

```python
# definindo um array 2d
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# escolhendo o eixo a ser permutado (linhas)
# somente alguns elementos são permutados
rgn.permuted(array_2d, axis=0)
```

```python
# definindo um array 2d
array_2d = np.arange(0, 20).reshape(4, 5)
array_2d
```

```python
# escolhendo o eixo a ser permutado (colunas)
# somente alguns elementos são permutados
rgn.permuted(array_2d, axis=1)
```

## Distribuições



- O NumPy fornece as principais distribuições para amostragem, a lista completa pode ser vista na documentação: https://numpy.org/devdocs/reference/random/generator.html#distributions

### `rgn.beta`

A distribuição beta gera uma quantidade de numeros menores e uma menor quantidade de numeros maiores.

```python
# a : alfa,
# b: beta (quanto maior beta, menor são os valores gerados)
# size: quantidade de numeros
rgn.beta(a=1, b=1, size=10)
```

```python
plt.hist(rgn.beta(a=1, b=100, size=10_000));
```

```python
# dividindo a distribuição em 100 intervalos
plt.hist(rgn.beta(a=1, b=100, size=10_000), bins=100);
```

### `rgn.binomial`

```python
# n = parametro da distribuição que deve ser maior que 1
# se n=10 os numeros gerados estarão entre 0 e 10
# p = probabilidade de cada numero ser sorteado
# size = quantidade numeros
rgn.binomial(n=10, p=0.5, size=10)
```

```python
plt.hist(rgn.binomial(n=10, p=0.5, size=100_000));
```

```python
# dividindo a distribuição em 100 intervalos
plt.hist(rgn.binomial(n=10, p=0.5, size=100_000), bins=100);
```

### `rgn.chisquare`

```python
# df: conceito da estatística que indica os graus de liberdade dever ser maior que zero
# quanto maior df, maior será os números randomicos gerados
rgn.chisquare(df=1, size=10)
```

```python
plt.hist(rgn.chisquare(df=5, size=100_000), bins=100);
```

### `rgn.exponential`

```python
rgn.exponential(scale=1, size=10)
```

```python
plt.hist(rgn.exponential(scale=1, size=100_000), bins=100);
```

### `rgn.gamma`

```python
rgn.gamma(shape=4, scale=2, size=10)
```

```python
plt.hist(rgn.gamma(shape=4, scale=2, size=100_000), bins=100);
```

### `rgn.logistic`

```python
rgn.logistic(loc=2, scale=5, size=10)
```

```python
plt.hist(rgn.logistic(loc=2, scale=5, size=100_000), bins=100);
```

### `rgn.normal`

```python
# loc: indica o centro da distribuição
# scale: desvio padrão dos numeros que serão ggerados
rgn.normal(loc=0, scale=1, size=10)
```

```python
plt.hist(rgn.normal(loc=0, scale=1, size=100_000), bins=100);
```

### `rgn.uniform`

```python
rgn.uniform(low=10, high=50, size=10)
```

```python
plt.hist(rgn.uniform(low=10, high=50,size=100_000), bins=100);
```
