### Comparando listas e arrays numpy - Vetorização

```python
# criando duas listas para tentar somar os elementos correspondentes
lista_1 = [1, 2, 3, 4]
lista_2 = [5, 6, 7, 8]

lista_3 = lista_1 + lista_2
lista_3 # o que ocorreu foi uma concatenação das duas listas
```

```python
# para somar os elementos correspondentes de uma lista
# precisamos fazer um for
lista_3 = []

for i in range(len(lista_1)):
    lista_3.append(lista_1[i] + lista_2[i])

lista_3
```

```python
# criando dois arrays numpy para tentar somar os elementos correspondentes
array_1 = np.array(lista_1)
array_2 = np.array(lista_2)

array_3 = array_1 + array_2
array_3 # foi feito a soma elemento por elemento
```

```python
# da mesma forma podemos fazer a subtração elemento por elemento
array_3 = array_1 - array_2
array_3
```

```python
# também podemos fazer a multiplicação elemento por elemento
array_3 = array_1 * array_2
array_3
```

```python
# podemos fazer a divisão elemento por elemento
array_3 = array_1 / array_2
array_3
```

```python
# podemos fazer o resto da divisão inteira elemento por elemento
array_3 = array_1 % array_2
array_3
```

```python
# podemos fazer a exponeciação elemento por elemento
# onde o primero array são as bases e o segundo são os expoentes
array_3 = array_1 ** array_2
array_3
```

### Comparando listas e arrays numpy - Broadcasting

```python
# definindo uma lista para somar elemento por elemento a uma constante
lista_1 = [1, 2, 3, 4]

lista_1 + 5 # TypeError: can only concatenate list (not "int") to list
```

```python
# devemos fazer um for para realizar a operação
lista_2 = []

for i in lista_1:
    lista_2.append(i + 5)

lista_2
```

```python
# definindo um array para somar elemento por elemento a uma constante
array_1 = np.array(lista_1)

array_2 = array_1 + 5
array_2
```

```python
# Também é possível fazer broadcasting com arrays de formato compativeis
array_1 = np.array(lista_1)
array_1
```

```python
array_2 = np.array([[5], [10]])
array_2
```

```python
array_2.shape # 2 linhas e 1 coluna (matriz)
```

```python
# observe que foi criado um novo array da seguinte forma:
# [[ 1+5,  2+5,  3+5,  4+5],
#  [ 1+10, 2+10, 3+10, 4+10]]
array_3 = array_1 + array_2
array_3
```

```python
# Porém nem sempre é possível fazer isso, as dimensões devem permitir:
array_4 = np.array([[5, 6],[10, 11]])
array_4
```

```python
array_4.shape # 2 linhas e 2 colunas (matriz)
```

```python
array_5 = array_1 + array_4
array_5 # ValueError: operands could not be broadcast together with shapes (4,) (2,2)
```

#### Arrays numpy de tamanho fixo

Abaixo vamos instanciar um array com 4 elementos.

```python
import numpy as np

array_a = np.array([0, 1, 2, 3])
```

Utilizando a função `np.append`, adicione o elemento `4` no final deste array. Lembre-se que este elemento deve estar "envelopado" em uma lista: `[4]`. Coloque o resultado em novo array chamado `array_b` e exiba o resultado.

```python
array_b = np.append(array_a, [4])
array_b
```

Agora, utilizando a função `id` do Python e o comparador `==`, mostre que os dois arrays têm endereços diferentes na memória, o que mostra que `array_b` é diferente de `array_a`.

💡 Isto ocorre porque, no NumPy, os arrays têm **tamanho fixo**. Se você adicionar elementos no array, um novo array é criado.

```python
id(array_a) == id(array_b)
```
