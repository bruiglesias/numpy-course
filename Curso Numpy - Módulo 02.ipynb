## ndarray e atributos

### Criando arrays a partir de listas

```python
import numpy as np
```

```python
# definindo um ndarray diretamente de uma lista
array_from_list = np.array([1, 2, 3])
array_from_list
```

```python
# definindo um ndarray a partir de uma variavel tipo lista
list_a = [1, 2, 3]
type(list_a)
```

```python
array_from_list = np.array(list_a)
array_from_list
```

```python
type(array_from_list)
```

```python
# O broadcasting pode ser feito normalmente
array_from_list + 5
```

### Criando arrays a partir de conjuntos

```python
# definindo uma ndarray a partir de uma variável do tipo conjunto
set_a = {1, 2, 3, 4, 5}
set_a
```

```python
type(set_a)
```

```python
array_from_set = np.array(set_a)

# o dtype do array criado é do tipo object
# o array possui um unico elemento que é o conjunto array_from_set[0] = {1, 2, 3, 4, 5}
array_from_set
```

```python
# O broadcasting não é suportado
# TypeError: unsupported operand type(s) for +: 'set' and 'int'

array_from_set + 5
```

### Criando arrays a partir de dicionário

```python
# definindo uma ndarray a partir de uma variável do tipo dicionário
dict_a = { 1: 'a', 2: 'b'}
dict_a
```

```python
type(dict_a)
```

```python
array_from_dict = np.array(dict_a)

# o dtype do array criado é do tipo object
# o array possui um unico elemento que é o conjunto array_from_dict[0] = {1: 'a', 2: 'b'}
array_from_dict
```

```python
# O broadcasting não é suportado
# TypeError: unsupported operand type(s) for +: 'dict' and 'int'

array_from_set + 5
```

### `size`

Retorna número total de elementos no array. Fornece a quantidade total de elementos, independentemente da forma (shape) do array

O atributo size é útil quando você precisa saber quantos elementos estão presentes no array sem precisar calcular manualmente a multiplicação de suas dimensões.


```python
array_1d
```

```python
array_1d.size # 3 elementos
```

```python
array_2d
```

```python
array_2d.size # 12 elementos
```

```python
array_3d
```

```python
array_3d.size # 24 elelmentos
```

### `ndim`

 É utilizado para retornar o número de dimensões (ou eixos) do array. Cada dimensão representa um nível de profundidade no array.

 é útil para entender a estrutura de um array e para aplicar operações que dependem do número de dimensões.

 É particularmente importante em áreas como ciência de dados, aprendizado de máquina, e processamento de imagens, onde os dados podem ter múltiplas dimensões.

```python
array_1d
```

```python
array_1d.ndim # 1 dimensão
```

```python
array_2d
```

```python
array_2d.ndim # 2 dimensões
```

```python
array_3d
```

```python
array_3d.ndim # 3 dimensões
```

### `shape`

É utilizado para retornar uma tupla de inteiros que indicam o tamanho do array em cada uma de suas dimensões. A tupla retornada representa o tamanho do array ao longo de cada uma de suas dimensões, fornecendo a forma (ou estrutura) do array.

É fundamental para entender a estrutura de um array e para realizar operações que dependem das dimensões específicas do array. A forma (shape) de um array é modificável, o que significa que você pode alterar a estrutura do array usando métodos como reshape.



```python
array_1d
```

```python
array_1d.shape
```

```python
array_2d
```

```python
array_2d.shape # 4 linhas e 3 colunas
```

```python
array_3d
```

```python
array_3d.shape # 2 matrizes (profundidade) com 4 linhas e 3 colunas
```

### `itemsize`

O atributo itemsize de um ndarray do NumPy é utilizado para retornar o tamanho (em bytes) de cada elemento do array. Este atributo é útil para entender o espaço de memória ocupado por cada elemento individualmente, o que pode ser importante para otimização de desempenho e gerenciamento de memória.

- O atributo itemsize é particularmente útil quando você precisa otimizar o uso de memória ou quando está lidando com grandes conjuntos de dados.

- Ele permite calcular o espaço total ocupado pelo array em memória ao multiplicar o valor de itemsize pelo número total de elementos (size).

```python
array_1d
```

```python
array_1d.itemsize # tamanho em bytes de cada elemento
```

```python
array_2d
```

```python
array_2d.itemsize # tamanho em bytes de cada elemento
```

```python
array_3d
```

```python
array_3d.itemsize # tamanho em bytes de cada elemento
```

### `nbytes`

O atributo nbytes de um ndarray do NumPy é utilizado para retornar o número total de bytes ocupados por todos os elementos do array. Este atributo é essencial para compreender a quantidade de memória utilizada pelo array em sua totalidade, o que é importante para otimização de desempenho e gerenciamento eficiente de memória, especialmente quando se lida com grandes conjuntos de dados.

```python
array_1d
```

```python
array_1d.nbytes # 3 * 8 = 24 bytes
```

```python
# outra forma de calcular é multiplicando o atributo 'size' por 'itemsize'
array_1d.size * array_1d.itemsize
```

```python
array_2d
```

```python
array_2d.nbytes # 4 * 3 * 8 = 96 bytes
```

```python
array_3d
```

```python
array_3d.nbytes # 2 * 4 * 3 * 8 = 192 bytes
```

### `dtype`

O atributo dtype de um ndarray do NumPy descreve o tipo de dados dos elementos contidos no array. Esse tipo de dados define a natureza dos elementos, como inteiros, floats, complexos, strings, etc., e especifica a precisão e o tamanho de cada elemento.

- O dtype é uma instância da classe 'numpy.dtype' e oferece várias propriedades e métodos para descrever e manipular o tipo de dados.

- Ele pode ser usado para criar arrays com tipos de dados específicos e pode ser alterado após a criação do array utilizando métodos como astype.

```python
array_1d
```

```python
array_1d.dtype # int64
```

```python
array_2d
```

```python
array_2d.dtype # int64
```

```python
array_3d
```

```python
array_3d.dtype # int64
```

```python
# modificando o tipo de dados dos elementos
array_32bits = np.array([1, 2, 3], dtype=np.int32)
array_32bits
```

```python
array_32bits.itemsize # 4 bytes
```

```python
# convertendo para outro tipo de dados (np.float32)
np.array(array_32bits, dtype=np.float32)
```

```python
np.array(array_32bits, dtype=np.float32).dtype
```

```python
# convertendo para outro tipo de dados (np.str_)
np.array(array_32bits, dtype=np.str_)
```

Estes são os atributos mais importantes do `ndarray`. Para uma lista completa, com alguns mais técnicos, visite a documentação: https://numpy.org/doc/stable/reference/arrays.ndarray.html#array-attributes.

### dtype

Agora vamos verificar como você pode instanciar um array explicitando o tipo de dado que será armazenado.

Abaixo você recebe uma lista, `list_d`, com 5 valores informados como inteiros.

```python
list_d = [1, 2, 3, 4, 5]

for i in list_d:
    print(i, type(i))
```

Partindo desta lista, instancie um array `array_d` convertendo seu `dtype` para `np.float16`.

Exiba o `dtype` deste array.

```python
array_d = np.array(list_d, dtype=np.float16)
array_d
```

```python
array_d.dtype
```

Agora instancie o mesmo `array_d` mas com `dtype` igual a `np.float32`. Exiba o `dtype`.

```python
array_d = np.array(list_d, dtype=np.float32)
array_d
```

```python
array_d.dtype
```

Novamente instancie `array_d`, mas agora com `dtype` igual a `np.str_`. Exiba o `dtype`.

```python
array_d = np.array(list_d, dtype=np.str_)
array_d
```

```python
array_d.dtype
```

Por último instancie `array_d` com `dtype` igual a `np.bool`.

Desta vez, exiba o próprio array e seu `dtype`.

```python
array_d = np.array(list_d, dtype=np.bool_)
array_d
```

```python
array_d.dtype
```

💡 Observe que todos os valores são expressos como `True`. Isto ocorre porque, para números, o Python interpreta `0` como `False`, e qualquer outro valor como `True`.
